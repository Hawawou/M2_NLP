```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def signal(t):
    sig = np.where(np.abs(t) <= 1, 2 - np.abs(t), 0)
    return sig
```

```python
t = np.linspace(-2, 2, 400)

plt.figure(figsize=(10,6))
plt.subplot(3, 1, 1)
plt.plot(t, signal(t), label='x(t)')
plt.title('Original Signal x(t)')
plt.grid(True)
plt.legend()
```

```python
# even part
def even_part(x, t):
    # x= signal, t = time array
    return 0.5 * (x(t) + x(-t))
```

```python
def odd_part(x, t):
    return 0.5 * (x(t) - x(-t))
```

```python
x_even = even_part(signal, t)
x_odd = odd_part(signal, t)

plt.figure(figsize=(10,6))
plt.subplot(3, 1, 2)
plt.plot(t, x_even, label='x_even(t)', color='g')
plt.title('Even Part of x(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3, 1, 2)
plt.plot(t, x_odd, label='x_odd(t)', color='r')
plt.title('Odd Part of x(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

## Unit step

```python
def unit_step(t):
    return np.where(t >= 0, 1, 0)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3, 1, 2)
plt.plot(t, unit_step(t), label='x(t)', color='r')
plt.title('Unit step signal')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x_even = even_part(unit_step, t)
x_odd = odd_part(unit_step, t)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3, 1, 2)
plt.plot(t, x_even, label='x_even(t)', color='g')
plt.title('Even Part of x(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3, 1, 2)
plt.plot(t, x_odd, label='x_odd(t)', color='r')
plt.title('Odd Part of x(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
