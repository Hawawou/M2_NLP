```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the corrected triangular signal x(t)
def signal(t):
    sig = np.where((t >= -1) & (t <= 0), t + 1, np.where((t > 0) & (t <= 2), 1 - t/2, 0))
    return sig

def time_reverse_and_scale(t):
    return signal(-t/2)

def time_scale_and_shift(t):
    return signal(2*(t + 2))

def time_reverse_and_shift(t):
    return signal(-(t+1))


# plot the signal
t = np.linspace(-4, 4, 400)
plt.figure(figsize=(12, 10))


plt.subplot(4, 1, 1)
plt.plot(t, signal(t), label='x(t)')
plt.title('Original Signal x(t)')
plt.plot(xlabel='t', ylabel='x(t)')
plt.grid()
plt.legend()

```

```python
# Time reverse and scale
plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(t, time_reverse_and_scale(t), label='x(-t/2)')
plt.title('Time Reversed and Scaled Signal x(-t/2)')
plt.plot(xlabel='t', ylabel='x(-t/2)')
plt.grid()
plt.legend()
```

```python
# Time scale and shift
plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(t, time_scale_and_shift(t), label='x(2(t + 2))')
plt.title('Time Scaled and Shifted Signal x(2(t + 2))')
plt.plot(xlabel='t', ylabel='x(2(t + 2))')
plt.grid()
plt.legend()
```

```python

plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.plot(t, time_reverse_and_shift(t), label='x(-(t+1))')
plt.title('Time Reversed and Shifted Signal x(-(t+1))')
plt.plot(xlabel='t', ylabel='x(-(t+1))')
plt.grid()
plt.legend()
```
