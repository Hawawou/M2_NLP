```python
df_reading = pd.read_csv('reading/original_features_reading.txt', sep=' ')
df_reading
```

```python
y_reading = df_reading['file'].str.slice(3,4)
print(y_reading)
```

```python
features_interview = df_interview.drop(columns='file')
features_interview
```

```python
features_reading = df_reading.drop(columns='file')
features_reading
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold


scaler = StandardScaler()
features_interview_scaled = scaler.fit_transform(features_interview)
features_reading_scaled = scaler.fit_transform(features_reading)

kf = KFold(n_splits=5)
for i, (train_index_interview, test_index_interview) in enumerate(kf.split(features_interview_scaled)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index_interview}")
    print(f"  Test:  index={test_index_interview}")
    X_train_interview = features_interview.iloc[train_index_interview]
    y_train_interview = y_interview.iloc[train_index_interview]
    X_test_interview = features_interview.iloc[test_index_interview]
    y_test_interview = y_interview.iloc[test_index_interview]

for i, (train_index_reading, test_index_reading) in enumerate(kf.split(features_reading_scaled)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index_reading}")
    print(f"  Test:  index={test_index_reading}")
    X_train_reading = features_reading.iloc[train_index_reading]
    y_train_reading = y_reading.iloc[train_index_reading]
    X_test_reading = features_reading.iloc[test_index_reading]
    y_test_reading = y_reading.iloc[test_index_reading]
```

```python
from sklearn.svm import SVC
from sklearn.metrics import balanced_accuracy_score


clf = SVC(kernel='linear')
clf.fit(X_train_interview, y_train_interview)
y_pred_interview = clf.predict(X_test_interview)
balanced_accuracy_score(y_test_interview, y_pred_interview)
```

```python
clf.fit(X_train_reading, y_train_reading)
y_pred_reading = clf.predict(X_test_reading)
balanced_accuracy_score(y_test_reading, y_pred_reading)
```
