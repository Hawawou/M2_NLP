# **Signal Decomposition into Even and Odd Components**

## **Objective**:
The goal of this notebook is to learn how to decompose any continuous-time signal into its even and odd components using Python. This exercise will involve defining two functions that compute the even and odd parts of a signal, respectively, and then visualizing the results using `matplotlib`.

### **Concepts Covered**:
1. **Even and Odd Functions**:
    - Any signal $ x(t) $ can be decomposed into two parts:
        - **Even component** $ x_e(t) $ which is symmetric: $ x_e(t) = x_e(-t) $
        - **Odd component** $ x_o(t) $ which is anti-symmetric: $ x_o(t) = -x_o(-t) $
    - Formula for even part: 
      $$
      x_e(t) = \frac{1}{2} \left[ x(t) + x(-t) \right]
      $$
    - Formula for odd part:
      $$
      x_o(t) = \frac{1}{2} \left[ x(t) - x(-t) \right]
      $$

2. **Visualization**:
    - After computing the even and odd parts, the signal and its components will be visualized.



## **Problem Statement**:
Given a signal \( x(t) \), decompose it into its even and odd parts. Plot the original signal, its even part, and its odd part. The signal \( x(t) \) can be any piecewise-defined function.

---

## **Instructions**:

1. **Define the Original Signal**:
    - We'll start by defining a function for $ x(t) $. The signal is piecewise linear:
      $$
      x(t) = 2 - |t| \quad \text{for} \quad -1 \leq t \leq 1, \quad \text{else} \quad 0
      $$

2. **Define Functions for Even and Odd Parts**:
    - Define two functions, `even_part(x, t)` and `odd_part(x, t)`, which compute the even and odd components based on the formulas.

3. **Plotting the Results**:
    - Use `matplotlib` to plot three figures:
      - The original signal $ x(t) $
      - The even part $ x_e(t) $
      - The odd part $ x_o(t) $

4. **Exercise**:
    - Modify the signal definition and observe how the even and odd parts change for different signals.

---

```python
## **Code Implementation**


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the example signal (piecewise linear)
def signal(t):
    """
    Define a piecewise linear signal that is 2-|t| for |t|<=1 and 0 elsewhere.
    """
    return np.where(np.abs(t) <= 1, 2 - np.abs(t), 0)
```

```python
# Define time range
t = np.linspace(-2, 2, 400)
```

```python
# Plot original signal, even part, and odd part
plt.figure(figsize=(10, 6))

# Plot the original signal x(t)
plt.subplot(3, 1, 1)
plt.plot(t, signal(t), label='x(t)')
plt.title('Original Signal x(t)')
plt.grid(True)
plt.legend()
```

```python
# Function to return the even part of a signal
def even_part(x, t):
    """
    Compute the even part of the signal.
    x: The signal function
    t: Time array
    """
    return 0.5 * (x(t) + x(-t))

# Function to return the odd part of a signal
def odd_part(x, t):
    """
    Compute the odd part of the signal.
    x: The signal function
    t: Time array
    """
    return 0.5 * (x(t) - x(-t))
```

```python
# Compute even and odd parts
x_even = even_part(signal, t)
x_odd = odd_part(signal, t)


# Plot the even part x_e(t)
plt.subplot(3, 1, 2)
plt.plot(t, x_even, label='x_e(t)', color='g')
plt.title('Even Part of the Signal x_e(t)')
plt.grid(True)
plt.legend()

# Plot the odd part x_o(t)
plt.subplot(3, 1, 3)
plt.plot(t, x_odd, label='x_o(t)', color='r')
plt.title('Odd Part of the Signal x_o(t)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

---

## **Detailed Code Explanation**:

### 1. **Even Part Function (`even_part`)**:
   - This function computes the even part of a signal \( x(t) \).
   - It takes two arguments:
     - `x`: the signal function \( x(t) \)
     - `t`: an array representing time
   - The function applies the formula \( x_e(t) = \frac{1}{2} \left[ x(t) + x(-t) \right] \) to return the even part.

### 2. **Odd Part Function (`odd_part`)**:
   - Similar to the even part, the `odd_part` function computes the odd part using the formula \( x_o(t) = \frac{1}{2} \left[ x(t) - x(-t) \right] \).

### 3. **Signal Definition**:
   - The `signal(t)` function defines a piecewise linear signal based on the absolute value of $ t \). The signal is non-zero between $ t = -1 $ and $ t = 1 $ and zero elsewhere.

### 4. **Plotting**:
   - We use `matplotlib` to plot three subplots:
     - The original signal $ x(t) $
     - The even part $ x_e(t) $
     - The odd part $ x_o(t) $
   - Each subplot includes a title and a grid for better visualization.
