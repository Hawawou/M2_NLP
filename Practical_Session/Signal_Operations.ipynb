### Homework Signal Operations in Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the corrected triangular signal x(t)
def signal(t):
    """
    A signal that is a line from -1 to 0 and another line from 0 to 2, otherwise 0.
    """
    return np.where((t >= -1) & (t <= 0), t + 1, np.where((t > 0) & (t <= 2), 1 - t / 2, 0))
```

```python
# Time vector for plotting
t = np.linspace(-4, 4, 400)
```

```python
# Plot the original signal and transformations
plt.figure(figsize=(12, 10))

# Original signal x(t)
plt.subplot(4, 1, 1)
plt.plot(t, signal(t), label='x(t)')
plt.title('Original Signal x(t)')
plt.grid(True)
plt.legend()
```

```python
# Define the transformations

# Time reversal and scaling x(-t/2)
def time_reverse_and_scale(t):
    return signal(-t / 2)
```

```python
# Time reversal and scaling x(-t/2)
plt.subplot(4, 1, 2)
plt.plot(t, time_reverse_and_scale(t), label='x(-t/2)')
plt.title('Time Reversal and Scaling: x(-t/2)')
plt.grid(True)
plt.legend()
```

```python
# Time scaling and shifting x(2(t+2))
def time_scale_and_shift(t):
    return signal(2 * (t + 2))
```

```python
# Time scaling and shifting x(2(t+2))
plt.subplot(4, 1, 3)
plt.plot(t, time_scale_and_shift(t), label='x(2(t+2))')
plt.title('Time Scaling and Shifting: x(2(t+2))')
plt.grid(True)
plt.legend()
```

```python
# Time reversal and shift x(-(t+1))
def time_reverse_and_shift(t):
    return signal(-(t + 1))


# Time reversal and shift x(-(t+1))
plt.subplot(4, 1, 4)
plt.plot(t, time_reverse_and_shift(t), label='x(-(t+1))')
plt.title('Time Reversal and Shifting: x(-(t+1))')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### Explanation of the Code:
1. **Original Signal**: 
   - The signal \( x(t) \) is defined as a triangular waveform that peaks at \( t = 0 \) and linearly decreases to 0 at \( t = -2 \) and \( t = 2 \).
   
2. **Time Reversal and Scaling** \( x(-t/2) \):
   - This operation involves reversing the signal (flipping it around the y-axis) and stretching it by a factor of 2. This is implemented by passing \( -t/2 \) as the argument to the `signal()` function.

3. **Time Scaling and Shifting** \( x(2(t+2)) \):
   - This is a combination of shifting the signal to the left by 2 units and compressing it by a factor of 2. The expression \( 2(t+2) \) achieves this transformation.

4. **Time Reversal and Shift** \( x(-t+1) \):
   - This operation first reverses the signal and then shifts it to the right by 1 unit. This is done by passing \( -(t-1) \) as the argument to the `signal()` function.

### Visualization:
- The program will generate four subplots:
  1. The **original signal** \( x(t) \) (triangular waveform).
  2. The **time-reversed and scaled** signal \( x(-t/2) \), which is stretched and flipped.
  3. The **time-scaled and shifted** signal \( x(2(t+2)) \), compressed and shifted to the left.
  4. The **time-reversed and shifted** signal \( x(-(t+1)) \), which is flipped and shifted to the right.
